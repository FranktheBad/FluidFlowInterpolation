This is an example of perfectly functional python code: 

```
word_list = ["this","is","an","example","of what","orginal","means"]

def remove_vowels_from(this_word):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    result = ""
    for i in range(len(this_word)):
        if this_word[i] not in vowels:
            result = result + this_word[i]
    return result        

result_list = []
for word in word_list:
    new_item = remove_vowels_from(word)
    result_list.append(new_item)

print(result_list)
```

The same results can be obtained with a more "pythonic" (but equally functional) code: 

```
import re
def rem_vowels(this_word):
    return (re.sub("[aeiouAEIOU]","",this_word))

result_list = [rem_vowels(words) for words in word_list]

print(result_list)
```
